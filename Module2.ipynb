## Task 1 - Plotting
* Plot the values of occupancy percentage versus number of weeks since release, for all movies in the dataset. 
* Plot the number of other releases in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.
* Plot the number of shows in the week versus occupancy percentage and weeks until the end of lifetime, for all movies.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('Data\\CinemaData.csv')
```

```python
data.columns
```

```python
plt.scatter(data.WeeksSinceRelease, data.OccPer, color = 'purple')
plt.xlabel('Occupancy Percentage')
plt.ylabel('Weeks Since Release')
plt.title('ALL MOVIES')
plt.show()

plt.scatter(data.OccPer, data.OtherReleasesInWeek, color = 'orange')
plt.ylabel('Number Of Other Releases In The Week')
plt.xlabel('Occupancy Percentage')
plt.title('OTHER RELEASES VS OCCUPANCY PERCENTAGE')
plt.show()

plt.scatter((data.Lifetime - data.WeeksSinceRelease), data.OtherReleasesInWeek, color = 'blue')
plt.ylabel('Number Of Other Releases In The Week')
plt.xlabel('Weeks until the end of lifetime')
plt.title('OTHER RELEASES VS WEEKS')
plt.show()

plt.scatter(data.OccPer, data.ShowsInWeek, color = 'red')
plt.ylabel('Number Of Shows In The Week')
plt.xlabel('Occupancy Percentage')
plt.title('SHOWS VS OCCUPANCY PERCENTAGE')
plt.show()

plt.scatter((data.Lifetime - data.WeeksSinceRelease), data.ShowsInWeek, color = 'black')
plt.ylabel('Number Of Shows In The Week')
plt.xlabel('Weeks until the end of lifetime')
plt.title('SHOWS VS WEEKS')
plt.show()
```
